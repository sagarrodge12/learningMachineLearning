```python
import matplotlib.pyplot as plt

#multiple points ploting
xpoints = [1,2,3,4,5]
ypoints = [10,20,30,40,50]

plt.plot(xpoints,ypoints)  # will plot the graph
plt.show()  # to show the plotting
```

```python
# two points ploting
xpoints = [0,6]
ypoints = [0,250]

plt.plot(xpoints,ypoints)
plt.show()
```

```python
# default x points. default x values are 0,1,2,3
# if one argmt give plot will consider that as ypoint
points = [10,20,30,40,50]

plt.plot(points)  # if you pass only one arg bu default it'll take that arg as ypoint and x point will run by default from 0 1 2 3 4 5..
plt.show()
```

```python
# format strings fmt  marker|line|color
# ms is marker size
# mfc is marker face color

ypoints = [3,8,1,10]

plt.plot(ypoints,marker = 'o', ms = 10, color = 'g', mfc ='b')
plt.show()
```

```python
# format argmt in one string
# """ is comment syntax
""" '-' solid line
    ':' dotted line
    '--' dashed line
    '-.' dash-dot line"""
ypoints = [3,8,1,10]
plt.plot(ypoints, 'o:k', linewidth='3')
# o represent marker
# : represent dotted line
# k represent black color
plt.show()
```

```python
# plotting multiple line in single graph
x1point = [1,2,3,4,5]
y1point = [10,20,30,40,50]

x2point = [1,50,6,20,17]
y2point = [10,20,30,40,50]

plt.plot(x1point,y1point)
plt.plot(x2point,y2point)
plt.show()
```

```python
# plotting multiple line in single graph
x1point = [1,2,3,4,5]
y1point = [10,20,30,40,50]

x2point = [1,50,6,20,17]
y2point = [10,20,30,40,50]

plt.plot(x1point,y1point,'o-.r',x2point,y2point,'*--y') # can add everything in plot
plt.show()
```

```python
# label and title to the line

xpoint = [1,50,6,20,17]
ypoint = [10,20,30,40,50]
plt.plot(xpoint,ypoint)
plt.xlabel("day")  # xlabel syntax for x axis
plt.ylabel("sales") # ylable syntax for y axis
plt.title("sales graph")
plt.show()
```

```python
#set font properties for title & label
# loc means location, 3 values are right, left, centre
xpoint = [1,50,6,20,17]
ypoint = [10,20,30,40,50]
plt.plot(xpoint,ypoint)

font1 = {'family':'serif','color':'blue','size':20} # created fonts
font2 = {'family':'serif','color':'darkred','size':15}
plt.xlabel("day", fontdict = font1)  # since font is dict keyword is fountdict
plt.ylabel("sales", fontdict = font2)
plt.title("sales graph", fontdict = font1, loc='right')
plt.show()


# no need to remember everything learning over hear is to know we can do
# something like this
```
