```python
import numpy as np
import matplotlib.pyplot as ply
## sklearn is lib where all algorithm are defined in this package ( linear, clusert, classification...Algorithm)
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

temp = [20,25,30,35,40]
sale = [13,21,25,35,38]

x= np.array([temp]).T  #linear regression methods expects a 2d array with rows and col
print(x)
print(x.shape)


```

```python
y=np.array(sale)
print(y)
print(y.shape)
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
y_pred = model.predict(x)
print(y_pred)
```

```python
#slope & intercept
print(model.coef_) #special variable definied with_
print(model.intercept_)
```

```python
#prediciting new value
newoutput = model.predict([[32]])
print(newoutput)
```

```python
mse = mean_squared_error(y,y_pred) # to evaulate the model and outdated
print(mse)
```

```python
import matplotlib.pyplot as plt
plt.scatter(temp, sale, marker="*", color ="r")
plt.plot(temp, y_pred, color="b")
plt.xlabel("Temperature")
plt.ylabel("Sales")
plt.show()
```

```python
# internally linearRegression working formula

from scipy import stats
slope, intercept, r, p, std_err = stats.linregress(temp, sale)
print('slope',slope)
print('intercept',intercept)
print('r',r)
print('p',p)
print('std_err',std_err)

#user defined fn
def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, temp))
print(mymodel)

plt.scatter(temp,sale)
plt.plot(temp, mymodel)
plt.show()

newsale = myfunc(45)
print(newsale)
```
